```python
import pandas as pd
import numpy as np

from pathlib import Path


data_dir = Path("Dataset")
```

## Load dataframes

```python
listens = pd.read_csv(data_dir / "listens.csv", index_col=False)
listens = listens[listens['uid'].map(listens['uid'].value_counts()) >= 1000] # we only want users with at least 1000 interactions
listens
```

```python
tot = {}
for usr in listens['uid'].unique():
    subset = listens[listens["uid"] == usr]
    if len(subset) > 1000:
        tot[usr] = len(subset)
len(tot)
```

```python
samples = 0

for usr, s in tot.items():
    samples += (s // 250) * 3 - 2
samples
```
